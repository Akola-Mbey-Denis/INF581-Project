# q-learning with different epsilon: we adapted q-learning algorithm so that we can study the influence of epsilon

# Experiment 1: to find the best num_seconds to train the agent using q-learing
## (1) single intersection
We trained our agent using q-learning method during 1000s, 5000s, 8000s, 10000s 
Then test the agend on the same road during 1000s,here is the result (average speed of vehicles) :

    * ql-Agent with 1000s : 3.15 m/s
    * ql-Agent with 5000s : 3.59 m/s
    * ql-Agent with 8000s : 3.82 m/s
    * ql-Agent with 10000s : 3.91 m/s

As we increase the training time, ql-Agent performes better in the testing environment. There no need to trained the agent more than 8000s, because there won't be significant improvement after that.( This can also be witnessed from the fig below). 




Besides, ql-agent learns quickly than the SARSA, In the SARSA experiment on the same single intersection road, we found that it take 10000s to train the agents in order to achieve an average speed of 3.87 for the testing. However, with the q-learning agent, it took only 8000s to train. This is because during ql-learning, the agent learns and updates it's q-table using epsilon-greedy policy and choose it's next action with greedy policy.

## (2) 2x2 grid
We trained our agent using q-learning method during 5000s, 8000s, 10000s, then test them during 1000s,here is the result (average speed of vehicles) :

     * ql-Agent with 5000s : 2.39 m/s
     * ql-Agent with 8000s : 2.61 m/s
     * ql-Agent with 10000s : 2.25 m/s
     
We see that ql-Agent trained by 8000s performed best, and it even performs better thant the SARSA agent trained by 10000s ( its result = 2.5 m/s). So genreally speaking, qlearning can train the agent faster than Sarsa. 

Here is the figure for the average at each time step during testing, there is a improvement of perfomence when the training time is increased from 5000s to 8000s. And from the figure, the agent with 8000s perfroms better than the other two agents, espcially in the testing step from 100s to 200s. 

This is an interesting results, because we are using a non-stationary state, and the 2x2 grid road is much more complicated than the single intersection. If we keep traing the agent during a very long time, it may keep updating its q-table for each states it encontered. Here is a risk, that during the training, the agent also learns q-value and took action accordingly using greedy policy. It may happens that if we keep training for a long time, the state of the road becomes stable, the q-value for each actions became simliar and the greedy policy became more random, and although it has learned a good q-value from previous steps, when the performed a new action, there is a risk that the agent took a more random action compared to the previous greedy policy, and the q-table for this state is update, it's a kind of "overfitting" for this problem.

To conclude: to train a q-learning agent in a complicated enviroment, increasing training time can increase its performence, but we should be careful with that, because q-learning uses greedy policy, it learns quickly, and there is the risk of "overfitting", wihich will lead to a bad performence at last.






# Experiment 2: to study the influence of epsilon on q-learing agent
## (1) single intersection

Here we fixed our training time to 8000s, and we chose epsilon as 0.1, 0.3, 0.5 to study its influence.
We still test our agent on the same environment during 1000s.

average speed of vehicles (m/s) are: 

    * ql-Agent with epsilon=0.1 : 3.94 m/s
    * ql-Agent with epsilon=0.3: 3.86 m/s
    * ql-Agent with epsilon=0.5 : 3.82 m/s

The fig shows average speed of vehicles at each time speed with different epsilon value. Actually we didn't see much difference visually, but from the mean value we can tell that the agent performed slightly better with a small epsilon. This is because if we decrease epsilon during training, the agent will be more likely to choose the best action, it has less chance to explore other actions, so it learns the best action of the given state better than higher epsilons. When epsilon is higher, during q-learning, the agent have more probability to explore a random action, so it does not always update the q-value for the best action. Our enviroment is nor a complicated one, without too much actions, using a small epsilon could help get a slightly better result.

To conclude, change the value of epsilon has a negeligant influence on this simple enviroment. 



## (2) 2x2 grid
Here we fixed our training time to 8000s, and we chose epsilon as 0.1, 0.3, 0.5 to study its influence.
We still test our agent on the same environment during 1000s.

average speed of vehicles (m/s) are: 

    * ql-Agent with epsilon=0.1 : 2.08 m/s
    * ql-Agent with epsilon=0.3:  2.26 m/s
    * ql-Agent with epsilon=0.5 : 2.61 m/s

At this 2x2 grid environment, we got a very different result compared to the previous single intersection test.

As we increse epsilon, we acturally have better performence. This is due to 2X2 grid is much more complicated than the single intersection. So basicly, we need more chance to choose differents actions in order to learn their q-value at a given stage. That is to say we need more exploration, as a result, increasing espislon could help us learn better under this complicated environment. 





# Conclusion:

(1) This is a non-stationary state enviroment, it does not have an actual terminal state, so we need to put an end to our learning process at after a proper training time.

(2) We've studied SARSA & Q-learning, generally speaking, q-learning could learn more quickly. Q-learning agent takes a smaller trainning number of seconds compared to sarsa agent. 

(3) For Q-learning agents, if we are using a complicated enviroment, like 2x2 grid, we shuold be award of the problem of "overfitting", we must choose a proper trainnig parameter to end the qlearnig process, before it's ruinning its performence.

(4) For Q-learning agents, we used epsilon greedy policy while training. For simple environments, epsilon has a negligeable influence on the performence of our agents. But for a complicated environment, It's better that we train our agent with a larger epsilon.

So it is really important to choose a proper training time for this "endless" non-stationary environment, and if the enviroment it complicated, it is recommended to use a higher epsilon value so that the agent could explore more while training, and this can actually improve its performence during testing.



# Q-learning algorithms

```python
def epsilon_greedy_policy(environment, q_table, state, action_space,epsilon):
    num_actions = environment.action_space.n
    if str(state) not in q_table:
        q_table[str(state)] = [0 for _ in range(num_actions)]
        action = environment.action_space.sample()
    else:

        if np.random.rand() < epsilon:
                action = environment.action_space.sample()
        else:
                action = np.argmax(q_table[str(state)])

    return action
```

```python
def greedy_policy(environment,q_table, state):
    num_actions = environment.action_space.n
    if str(state) not in q_table:
        q_table[str(state)] = [0 for _ in range(num_actions)]
        action = environment.action_space.sample()
    else:
        action = np.argmax(q_table[str(state)])
    return action
```

```python
q_array_history = []
alpha_history = []

def qlearning(environment, alpha=0.1, alpha_factor=0.9995, gamma=0.99, epsilon=0.5):

    #initial states:
    s  = environment.reset()
#     num_states = environment.observation_space.n
    num_actions = environment.action_space.n
    # Initial Q table

    q_array = {str(s): [0 for _ in range(num_actions)]}
    if str(s) not in q_array:
            q_array[str(s)] = [0 for _ in range(num_actions)]

    q_array_history.append(q_array.copy())
    alpha_history.append(alpha)

    # Update alpha
    if alpha_factor is not None:
        alpha = alpha * alpha_factor

    end = False

    while not end:
        a = epsilon_greedy_policy(environment,q_array,str(s),environment.action_space, epsilon)
        s_, r_,  terminated, truncated, info = environment.step(a)
        a_ = greedy_policy(environment,q_array,str(s_))
        if str(s) not in q_array:
            q_array[str(s)] = [0 for _ in range(num_actions)]
        if str(s_) not in q_array:
            q_array[str(s_)] = [0 for _ in range(num_actions)]

        q_array[str(s)][a]  = q_array[str(s)][a]+  alpha*(r_ + gamma*q_array[str(s_)][a_]-q_array[str(s)][a])

        s = s_
        
        end  = terminated or truncated

    return q_array
```

```python
def evaluate_policy(q_array,net_file,route_file,output,num_seconds = 1000):
    
    reward_list = []
    num_seconds = num_seconds
    env = gym.make('sumo-rl-v0',
                    net_file=net_file,
                    route_file=route_file,
                    out_csv_name='outputs/single-intersection/output.csv',
                    use_gui=False,
                    num_seconds=num_seconds)
    state, info = env.reset()
    done = False
    while not done:
        action = greedy_policy( env,q_array,state)
        state, reward,  terminated, truncated, info = env.step(action)
        reward_list.append(reward)
        done = terminated or truncated
    env.save_csv(output,1)
    env.close()
```
